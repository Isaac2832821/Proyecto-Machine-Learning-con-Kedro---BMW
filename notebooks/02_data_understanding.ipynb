```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo para gráficos
sns.set(style="whitegrid", palette="muted", font_scale=1.2)
```

```python
# Cargar datos desde la carpeta raw
df = pd.read_csv("../data/01_raw/bmw_sales.csv")

print("Dimensiones:", df.shape)
print("\nColumnas:", df.columns.tolist())
df.head()
```

```python
# Info general del dataset
df.info()

# Resumen estadístico
df.describe(include="all")
```

```python
# Info general del dataset
df.info()

# Resumen estadístico
df.describe(include="all")
```

```python
print("Valores nulos por columna:\n", df.isnull().sum())
print("\nCantidad de duplicados:", df.duplicated().sum())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price_USD"], bins=30, kde=True, color="blue")
plt.title("Distribución de precios (USD)")
plt.xlabel("Precio (USD)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Sales_Volume"], bins=30, kde=True, color="orange")
plt.title("Distribución del volumen de ventas")
plt.xlabel("Unidades vendidas")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Fuel_Type", y="Price_USD", data=df)
plt.title("Precio de autos según tipo de combustible")
plt.xlabel("Tipo de combustible")
plt.ylabel("Precio (USD)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Transmission", y="Price_USD", data=df)
plt.title("Precio según tipo de transmisión")
plt.xlabel("Transmisión")
plt.ylabel("Precio (USD)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación entre variables numéricas")
plt.show()
```
